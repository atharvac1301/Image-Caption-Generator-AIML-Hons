```python
from transformers import pipeline
from PIL import Image
import matplotlib.pyplot as plt


# Load pre-trained image captioning pipeline
caption_generator = pipeline("image-to-text", model="Salesforce/blip-image-captioning-large")

# Load your image
file_path = "image-2.png"
image = Image.open(file_path)

# Generate caption
caption = caption_generator(image)

# Display Image
plt.imshow(image)
plt.axis('off')
plt.show()

# Print result
print("Caption:", caption[0]['generated_text'])

```
